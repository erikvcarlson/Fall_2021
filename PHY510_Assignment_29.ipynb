```python
import math 
import numpy as np
import sympy as sp
const = 1/(2 * math.pi)
n = 25
l = 50
v = np.linspace(-5,5,l)
f_v = []
x=sp.Symbol('x') 
a = 0
for i in range(-n,n):
    a += sp.exp(1j*i*x)

g = a.subs('x',v)


for x in range(len(v)): 
    f_v.append(sp.N(a.subs('x',v[x])))

    
x = [ sp.re(ele) for ele in f_v]

x = 1/(2*math.pi)*np.array(x)

import matplotlib.pyplot as plt
plt.plot(v,x)
```
