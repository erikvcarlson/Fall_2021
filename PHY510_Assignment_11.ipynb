```python
from math import exp
import numpy as np
import matplotlib.pyplot as plt

def y_t(t,t1,t2):
    return exp(-t/t1) + exp(-t/t2)

def f_t1_t2(list_1):
    t1 = list_1[0]
    t2 = list_1[1]
    sum_start = np.zeros(4)
    for j in range(1,4):
        sum_start[j] = (y_t(j,t1,t2) - y_t(j,2,3))**2
    return np.sum(sum_start)


delta = 0.01
x = np.arange(delta, 10, delta)
y = np.arange(delta, 10, delta)
X, Y = np.meshgrid(x, y)
z = np.zeros([len(X),len(Y)])

for i in range(0,len(X)):
    for j in range(0,len(Y)):
        pass_var = [X[i,j],Y[i,j]]
        z[i,j] = f_t1_t2(pass_var)


        
#from scipy import optimize
#gamma = optimize.brent(f_t1_t2,brack=[(1,2),(1,2)])
gamma = 0.1
#print(np.unique(z))
grad = np.gradient(z)    
    
print(grad[0])
fig = plt.figure(figsize=(6,5))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 
cp = ax.contour(x, y, z,[0.01,0.03,0.1,0.2,0.4,0.6,0.8,1,1.2])
ax.clabel(cp, inline=True, 
          fontsize=10)
ax.set_title('Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()

        
        
        
```

```python
import numpy as np
import math 

gr = (math.sqrt(5) + 1) / 2

def gss_x(a, b,y, tol=1e-10):
    c = b - (b - a) / gr
    d = a + (b - a) / gr
    while abs(b - a) > tol:
        list_1 = [c,y]
        list_2 = [d,y]
        if f_t1_t2(list_1) < f_t1_t2(list_2):
            b = d
        else:
            a = c
        c = b - (b - a) / gr
        d = a + (b - a) / gr

    return (b + a) / 2

def gss_y(a, b,x, tol=1e-10):
    c = b - (b - a) / gr
    d = a + (b - a) / gr
    while abs(b - a) > tol:
        list_1 = [x,c]
        list_2 = [x,d]
        if f_t1_t2(list_1) < f_t1_t2(list_2):
            b = d
        else:
            a = c
        c = b - (b - a) / gr
        d = a + (b - a) / gr

    return (b + a) / 2


a = gss_x(1,4,2)
b = gss_y(1,4,a)


print([a,b])
b = gss_y(1,4,3)
a = gss_x(1,3,b)
print([a,b])
```
