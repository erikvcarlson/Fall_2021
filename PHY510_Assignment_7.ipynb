```python
import numpy
import scipy
import math
from mpmath import *
import pandas as pd
from scipy.special import hyp2f1
```

```python
def recursive_relation(i,a):
    if i == 0:
        return(mp.mpf(math.log(1+1/a)))
    else:
        return 1/i - a*recursive_relation(i-1,a)

def value_generator(dps,a):
    rounded_data_list = []
    mp.dps = dps
    y_0 = mp.mpf(math.log(1+1/a))
    for i in range(0,15):
        y_n = mp.mpf(hyp2f1(1,1+i,2+i,-a**-1)/(a*(1+i)))
        y_n_recu = recursive_relation(i,a)
        y_store = mp.mpf(round(y_n - y_n_recu,5))
        print(y_store)
        rounded_data_list.append(y_store)
    return rounded_data_list

#starting at the same value as the hyp2f1 function and diverging
dps = [5,10,15,20]
a = 10
dl = []
for i in range(0,len(dps)):
    dl.append(value_generator(dps[i],a))
    print('------- \n')
    
```

```python
dps = [5,10,15,20]
a = 10
dl = []

def value_generator_backwards(dps,a):
    rounded_data_list = []
    mp.dps = dps
    y_0 = mp.mpf(math.log(1+1/a))
    for i in range(15,0,-1):
        y_n = mp.mpf(hyp2f1(1,1+i,2+i,-a**-1)/(a*(1+i)))
        y_n_recu = recursive_relation(i,a)
        y_store = mp.mpf(round(y_n - y_n_recu,5))
        print(y_store)
        rounded_data_list.append(y_store)
    return rounded_data_list


#starting at the diverged value of the recursion relation and converging on the hyp2f1 function
for i in range(0,len(dps)):
    dl.append(value_generator_backwards(dps[i],a))
    print('------- \n')
```
