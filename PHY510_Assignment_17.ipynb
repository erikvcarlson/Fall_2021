```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#plotting a histogram of of the sum of two random matrices


n = 50000
a = np.random.rand(n)
b = np.random.rand(n)
c = a + b
vari_list = [np.var(a),np.var(b)]
plt.hist(c,bins=50)
plt.show()


```

```python
#fitting our gaussian to a histrogram of the probability density of the sum of our two random matrices 
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

g = plt.hist(c, density = True,bins=50)


mu = np.mean(c)
variance = sum(vari_list)
sigma = math.sqrt(sum(vari_list))
x = np.linspace(min(c), max(c), 50)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()

```

```python
def array_prob(m):
    n = 50000
    c = np.zeros(n)
    vari_list = []
    for i in range(0,m):
        a = np.random.rand(n)
        vari_list.append(np.var(a))
        c = c + a
    return [c,vari_list]
[c,vari_list] = array_prob(50)

#doing the same steps as the first cell, this time with an arbitrary number of arrays
plt.hist(c,bins=50)
plt.show()
d = plt.hist(c,density = True,bins=50)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

#we find the mean of the whole and the variance of the parts and use that coupled with an idealized
#gaussian function which is plotted below
mu = np.mean(c)
variance = sum(vari_list)
sigma = math.sqrt(sum(vari_list))
x = np.linspace(min(c), max(c), 50)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
#we can find the difference of the values of the probability densities of our approximated gaussian and our actual
#gaussian below
diff = d[0] - stats.norm.pdf(x, mu, sigma)
plt.plot(range(0,50),diff)
```
